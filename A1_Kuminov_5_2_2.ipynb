<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
#Ваш код
import os
path = "data\stock"
files = os.listdir(path) #- список файлов в папке file_path
dfStock = pnd.DataFrame()
names = []
for file in files:
    name = file[::-1]
    name = name[4:]
    name = name[::-1]
    tempdf = pnd.read_csv("data\stock\{}".format(file), parse_dates=True)#- загрузка из файла в dataframe
    names.append(name)
    templist = []
    for i in range(len(tempdf)):
        templist.append(name)
    tempdf['Name'] = templist
    dfStock = pnd.concat([tempdf,dfStock],ignore_index = True)
tempdf = dfStock[["Name","Close"]]# - нужна только цена закрытия
dfStock
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
#Ваш код
correlationdf = pd.DataFrame()
for name in names:
    values = pd.DataFrame(data = list(tempdf[tempdf.Name == name].Close), columns = [name])
    correlationdf = pd.concat([correlationdf,values],ignore_index = True,axis = 1)
correlationdf.columns = names
m = correlationdf.corr()
```
