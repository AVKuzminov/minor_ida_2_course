```python
import statsmodels
import scipy as sc
import numpy as np
import pandas as pnd
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.graphics.regressionplots import plot_leverage_resid2
import matplotlib.pyplot as plt
%pylab inline
```

```python
raw = pnd.read_csv("beauty.csv",sep=';',index_col=False)
raw.head()
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(raw[["wage","exper","educ","looks"]], figsize=(15, 15))
pylab.show()
```

```python
print("union: \n{0}".format((raw["union"].value_counts())))
print("goodhlth: \n{0}".format((raw["goodhlth"].value_counts())))
print("black: \n{0}".format((raw["black"].value_counts())))
print("female: \n{0}".format((raw["female"].value_counts())))
print("married: \n{0}".format((raw["married"].value_counts())))
print("service: \n{0}".format((raw["service"].value_counts())))
```

```python
plt.hist(raw.wage,bins=20)
plt.hist(np.log(raw.wage), bins=20)
plt.show()
```

```python
raw = raw[raw.wage < 77] # 
```

```python
plt.hist(raw.looks)
```

```python
raw["LooksWorseThan3"] = raw["looks"].apply(lambda x: 0 if (x>=3) else 1)
```

```python
raw["LooksBetterThan3"] = raw["looks"].apply(lambda x: 1 if (x>=3) else 0)
raw = raw.drop("looks",axis=1)
raw.tail()
```

```python
ml = smf.ols("wage ~ exper + union + goodhlth + black + female + married + service + educ + LooksWorseThan3 + LooksBetterThan3"
             , data = raw)
ml = ml.fit()
print(ml.summary())
```

```python
mlog = smf.ols("np.log(wage) ~ exper + union + goodhlth + black + female + married + service + educ + LooksWorseThan3 + LooksBetterThan3"
             , data = raw)
mlog = mlog.fit()
print(mlog.summary())
```

```python
plt.figure(figsize = (10,10))
plt.subplot(121)
sc.stats.probplot(ml.resid, dist="norm",plot=pylab)
plt.subplot(122)
np.log(ml.resid).plot.hist()
pylab.show()
```

```python
plt.figure(figsize = (10,10))
plt.subplot(121)
sc.stats.probplot(mlog.resid, dist="norm",plot=pylab)
plt.subplot(122)
np.log(mlog.resid).plot.hist()
pylab.show()
```
